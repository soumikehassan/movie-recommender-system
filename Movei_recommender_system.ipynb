```python
#make list
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
#Define a function for this process
def convert(genres_list):
  L =[]
  for i in ast.literal_eval(genres_list):
    L.append(i['name'])
  return L
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
#Again Define function
def convert_2(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_2)
```

```python
movies['cast'][0]
```

```python
movies['crew'][0]
```

```python
def find_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(find_director)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in  x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in  x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in  x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in  x])
```

```python
movies['tags']=movies['overview']+ movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new_movies = movies[['movie_id','title','tags']]
```

```python
new_movies['tags']=new_movies['tags'].apply(lambda x:" ".join(x))
```

```python
#convert all text into lower case
new_movies['tags']=new_movies['tags'].apply(lambda x:x.lower())
```

```python
#Text Vectorization
# Bag of words
# tf idf
# remove unwanted word like as in,and ,or ,is
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_movies['tags']).toarray()
```

```python
# Import the PorterStemmer class
from nltk.stem.porter import PorterStemmer

# Create a PorterStemmer object
ps = PorterStemmer()

# Define the stem function correctly
def stem(text):
    y = []
    for word in text.split():
        y.append(ps.stem(word))
    return " ".join(y)

# Apply the stem function to the 'tags' column of the new_movies DataFrame
new_movies['tags'] = new_movies['tags'].apply(stem)
```

```python
#here we use cosine distance ,not equildien distance
# cosine distance means that we calculate the angel between two movies
# for high dimentional data equilidien distance not work perfectly
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
def recommend(movie):
    index = new_movies[new_movies['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_movies.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
